```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
# The bootstrap method
# Part 1: Homework Hours Problem
hours = [2, 2, 3, 3, 5]
r = 1000

# Bootstrap sampling for medians, variance, and bias
bootstrap_medians = [np.median(np.random.choice(hours, len(hours), replace=True)) for _ in range(r)]
variance_median = np.var(bootstrap_medians)
bias_median = np.mean(bootstrap_medians) - np.median(hours)

plt.hist(bootstrap_medians, bins=20, edgecolor='k', alpha=0.7)
plt.title('Bootstrap Distribution of Sample Median')
plt.xlabel('Median Hours')
plt.ylabel('Frequency')
plt.show()

# Part 2: GRE Scores Problem
gre_scores = [5.22, 4.74, 6.44, 7.08, 4.66, 5.34, 4.22, 4.80, 5.02, 6.55, 4.18, 4.64, 6.00, 4.12, 5.30, 5.64]
bootstrap_means_gre = [np.mean(np.random.choice(gre_scores, len(gre_scores), replace=True)) for _ in range(10000)]
estimated_average_gre = np.mean(bootstrap_means_gre)

plt.hist(bootstrap_means_gre, bins=20, edgecolor='k', alpha=0.7)
plt.title('Bootstrap Distribution of Sample Mean (GRE Scores)')
plt.xlabel('Mean GRE Score')
plt.ylabel('Frequency')
plt.show()

# Part 3: Boston Housing Dataset Problem
data = fetch_openml(name="boston", version=1, as_frame=True).frame
medv = data['MEDV'].values

mean_medv = np.mean(medv)
standard_error_formula = np.sqrt(np.var(medv, ddof=1) / len(medv))
bootstrap_means_medv = [np.mean(np.random.choice(medv, len(medv), replace=True)) for _ in range(10000)]
standard_error_bootstrap = np.std(bootstrap_means_medv)

print(f"Part 1 (Homework): Variance of Median = {variance_median}, Bias of Median = {bias_median}")
print(f"Part 2 (GRE): Estimated Average GRE Score = {estimated_average_gre}")
print(f"Part 3 (Boston Housing): Mean MEDV = {mean_medv:.2f}, SE (Formula) = {standard_error_formula:.2f}, SE (Bootstrap) = {standard_error_bootstrap:.2f}")
```
