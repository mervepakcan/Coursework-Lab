```python
import numpy as np
import matplotlib.pyplot as plt

#Generating 1000 samples, each containing 1000 observations
n=1000 #each sample consist of 1000 observations
t=1000 #the number of samples to generate

# Generate random samples
sample_means = [] # List to store the sample means
for _ in range(t):
    sample = np.random.uniform(0, 1, n)  # Drawing n random samples from U(0, 1)
    sample_mean = np.mean(sample)  # Computing the sample mean
    sample_means.append(sample_mean) # Append the calculated mean to the list

# Plot histogram
plt.hist(sample_means, bins=30, edgecolor='k', alpha=0.7)
plt.title('Histogram of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()

#Observations
# Due to CLT,the distribution of sample means approximates a normal distribution, despite the underlying data being uniformly distributed.
# The histogram is center around 0.5, which is the mean of the uniform distribution
# Since the sample size is large (n=1000), the sample means cluster tightly around 0.5, resulting in a narrow distribution.
```
