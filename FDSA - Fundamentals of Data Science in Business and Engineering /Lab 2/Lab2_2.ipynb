```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2  # Importing the chi-squared distribution function from scipy

# Parameters for the simulation
n = 3  # number of observations in each sample (degrees of freedom = n - 1 = 2)
num_samples = 1000  # Total number of chi-squared samples to generate

# Generating random samples and computing chi-squared values
chi_squared_values = []  # List to store the computed chi-squared values

# Loop to generate samples and compute their chi-squared values
for _ in range(num_samples):
    z = np.random.normal(0, 1, n)  # n samples from N(0, 1)
    chi_squared_value = np.sum(z**2)  # The sum of the squares of the samples
    chi_squared_values.append(chi_squared_value) # Appending the chi-squared value to the list

plt.hist(chi_squared_values, bins=30, edgecolor='k', alpha=0.7, density=True, label='Simulated')
# The histogram visualizes the distribution of the simulated chi-squared values
# density=True normalizes the histogram so the area sums to 1.

# Adding the theoretical chi-squared distribution using scipy
x = np.linspace(0, 10, 100)  # Generating 100 points evenly spaced between 0 and 10
# Calculating the chi-squared PDF for 2 degrees of freedom
chi_squared_pdf = chi2.pdf(x, df=n-1)

plt.plot(x, chi_squared_pdf, label='Chi-squared (df=2)', color='red')  # Theoretical chi-squared curve

# Plot configuration
plt.title('Simulating a xÂ² Distribution')  # Title of the plot
plt.xlabel('Sum of Squares')  # Label for the x-axis indicating the chi-squared values
plt.ylabel('Density')  # Label for the y-axis indicating the probability density
plt.legend()
plt.show()

#Observations
# The histogram displays a right-skewed distribution, which is typical with low degrees of freedom, such as df=2.
# The results has not symmetrical shape.
# A right-skewed histogram features a longer tail on the right side, with the peak shifted towards the left. This indicates that smaller values are more frequent in the dataset, while larger values are less common. It is also referred to as a positively skewed histogram.
# The red line represents the theoretical chi-squared distribution with 2 df and closely match the histogram, showcasing the accuracy of the simulation.
# The histogram and theoretical curve peak around 1-2 with most values concentrated near zero, extending into a long tail on the right.
```
